# Statistics Advance - 1

**1)  What is a random variable in probability theory?**

-> A random variable is a function that assigns numerical values to the outcomes of a random experiment. It represents uncertain events in a mathematical way and allows us to analyze probabilities associated with different outcomes.

For example, if you roll a die, the number that appears on the top face (1 to 6) is a random variable.

**2) What are the types of random variables?**

-> **There are two main types of Random Variables :**

* **Discrete Random Variable:** Takes a countable number of distinct values (e.g., integers).

Example: The number of heads when flipping a coin 3 times (0, 1, 2, or 3).
*  **Continuous Random Variable:** Takes an uncountable number of values (e.g., real numbers).

Example: The exact height of a person (e.g., 5.72 feet).

**3)  What is the difference between discrete and continuous distributions?**

-> A **discrete distribution** deals with countable values, like the number of heads in a coin toss. It uses a **Probability Mass Function (PMF)** to assign probabilities to specific values, which can be nonzero.

Example-Rolling a die (1, 2, 3, 4, 5, 6).

A **continuous distribution** deals with uncountable values, like a person's height. It uses a **Probability Density Function (PDF)**, and probabilities are given over intervals since the probability of any single value is zero.

Example- Heights of people (e.g., 5.1 ft, 5.5 ft, etc.).

**4) What are probability distribution functions (PDF)?**

-> A Probability Distribution Function (PDF) describes how probabilities are distributed over the values of a random variable. There are two types:
* For discrete variables, the function is called a Probability Mass Function (PMF), which assigns a probability to each possible outcome.
* For continuous variables, the function is called a Probability Density Function (PDF), which represents the likelihood of the variable taking on a value within an interval.

**5) How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?**

-> Difference Between CDF and PDF :
* A Cumulative Distribution Function (CDF) gives the probability that a random variable is less than or equal to a given value.
* A Probability Distribution Function (PDF) gives the probability (discrete case) or probability density (continuous case) for a specific value.

Mathematically, the CDF is the integral (summation in discrete cases) of the PDF.

**6) What is a discrete uniform distribution?**

-> A Discrete Uniform Distribution is a probability distribution where all outcomes in a finite set are equally likely.
* If there are 'n' possible values, each has a probability of 1/n.

Example: Rolling a fair die, where each face (1 to 6) has an equal probability of 1/6.

**7) What are the key properties of a Bernoulli distribution?**

-> A Bernoulli distribution represents a random experiment with exactly two possible outcomes (success or failure).
* Parameter: p (probability of success).
* Mean (Expected Value) : E[X] = p
* Variance: Var(X) : p(1-p)

Example: Tossing a coin (Heads = 1, Tails = 0).

**8) What is the binomial distribution, and how is it used in probability?**

-> A Binomial distribution models the number of successes in 𝑛 independent Bernoulli trials.

**Formula :** $$ P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k} $$

where:
* 𝑛 is the total number of trials,
* 𝑘 is the number of successful trials,
* 𝑝 is the probability of success in each trial.

**Mean:** 𝑛𝑝

**Variance:** 𝑛𝑝(1−𝑝)

**Used in:** Quality control, genetics, and decision-making processes.

**9)  What is the Poisson distribution and where is it applied?**

-> The Poisson distribution models the probability of a given number of events occurring in a fixed interval of time or space.

**Formula :**  $$ P(X = k) = \frac{e^{-\lambda} \lambda^k}{k!} $$

where:
* 𝜆 is the mean number of occurrences,
* 𝑘 is the number of occurrences,
* 𝑒 is Euler's number (~2.718).

**Mean and Variance:** 𝜆

**Used in:** Call centers (number of calls per hour), traffic flow, and rare event modeling.

**10) What is a continuous uniform distribution?**

-> A continuous uniform distribution is a probability distribution in which all values within a given interval are equally likely to occur. It is defined by two parameters:
* a: The lower bound (minimum value)
* b: The upper bound (maximum value)

The probability density function (PDF) of a continuous uniform distribution is:

$$ f(x) =
\begin{cases}
\frac{1}{b - a}, & a \leq x \leq b \\
0, & \text{otherwise}
\end{cases} $$

The cumulative distribution function (CDF) is:

$$ F(x) =
\begin{cases}
0, & x < a \\
\frac{x - a}{b - a}, & a \leq x \leq b \\
1, & x > b
\end{cases} $$

The mean and variance are given by:
$$ E[X] = \frac{a + b}{2} $$

$$ Var(X) = \frac{(b - a)^2}{12} $$

**11) What are the characteristics of a normal distribution?**

->A Normal Distribution (Gaussian distribution) is a continuous probability distribution that follows a bell-shaped curve.

**Key Characteristics:**
* Symmetry: The distribution is symmetric about the mean.
* Mean = Median = Mode: All three are equal and located at the center of the distribution.
* Unimodal: It has a single peak.
* Asymptotic: The tails extend infinitely without touching the x-axis.
* Empirical Rule (68-95-99.7 Rule):
  * 68% of data falls within 1 standard deviation (σ) of the mean.
  * 95% falls within 2 standard deviations.
  * 99.7% falls within 3 standard deviations.

**12) What is the standard normal distribution, and why is it important?**

-> A Standard Normal Distribution is a special case of a normal distribution where:
* Mean (μ) = 0
* Standard Deviation (σ) = 1

**Importance:**
* Allows standardization of normal distributions, making it easier to compare different datasets.
* Used in Z-scores and probability calculations.
* Helps in hypothesis testing and confidence intervals.

**13) What is the Central Limit Theorem (CLT), and why is it critical in statistics?**

-> The Central Limit Theorem (CLT) states that:

The sampling distribution of the sample mean will approximate a normal distribution, regardless of the original population distribution, if the sample size is sufficiently large.

It is Critical because:
* Allows us to use normal distribution techniques (Z-tests, t-tests) even when the population is not normally distributed.
* Enables statistical inference, making hypothesis testing and confidence intervals possible.
* Works best when sample size n≥30.

**14)  How does the Central Limit Theorem relate to the normal distribution?**

* CLT explains why many real-world distributions are approximately normal.
* Even if a population is skewed or non-normal, its sample means will form a normal distribution for large n.
* This justifies using the normal approximation in statistical methods.

**15) What is the application of Z statistics in hypothesis testing?**

-> Z-statistics (Z-tests) are used when:
* The population follows a normal distribution.
* The sample size is large (n≥30).
* The population standard deviation (σ) is known.

**Uses in Hypothesis Testing:**
* One-sample Z-test: Compares a sample mean to a known population mean.
* Two-sample Z-test: Compares two independent sample means.
* Proportion Z-test: Tests differences in population proportions.

**16) How do you calculate a Z-score, and what does it represent?**

-> A Z-score measures how many standard deviations a data point is from the mean.

Formula:
$$ Z= (X−μ )/σ  $$
where:
X = observed value
μ = mean
σ = standard deviation

Interpretation:
* Z = 0 → The value is at the mean.
* Z > 0 → The value is above the mean.
* Z < 0 → The value is below the mean.
* |Z| > 2 → The value is unusual.
* |Z| > 3 → The value is an outlier.

**17  What are point estimates and interval estimates in statistic?**

-> Point Estimates vs. Interval Estimates

**Point Estimate:** A single value used to estimate a population parameter (e.g., sample mean).

**Interval Estimate:** A range of values (confidence interval) that likely contains the population parameter.

**18) What is the significance of confidence intervals in statistical analysis?**

-> A Confidence Interval (CI) provides a range of values that likely contains the population parameter.

It's Important becasuse:
* Helps in decision-making with a margin of error.
* More informative than a point estimate.
* Common confidence levels: 90%, 95%, 99%.

Formula for CI (when population standard deviation is known):

$$ CI = \bar{x} \pm Z_{\alpha/2} \times \frac{\sigma}{\sqrt{n}} $$

where, $$ Z_{\alpha/2} $$ is the Z-score for the desired confidence level.

**19)  What is the relationship between a Z-score and a confidence interval?**

-> Relationship Between Z-Score and Confidence Interval:

A Z-score is used to determine the margin of error in a confidence interval.
Higher confidence levels require larger Z-scores (wider intervals).
Example:
* 90% CI → Z = 1.645
* 95% CI → Z=1.96
* 99% CI → Z=2.576

**20) How are Z-scores used to compare different distribution.?**

-> Z-Scores Are Used to Compare Different Distributions By:
* Standardizes different distributions for direct comparison.
* Allows comparison of values from different units and scales.
* Helps in detecting outliers and extreme values across datasets.

**21)  What are the assumptions for applying the Central Limit Theorem?**

-> Assumptions for applying the Central Limit Theoremare:
* **Random Sampling:** Samples must be randomly drawn.
* **Independent Observations:** Each sample point should be independent.
* **Sample Size** n is Large: n≥30 is typically sufficient.
* **Finite Variance:** The population should have a finite variance.

**22) What is the concept of expected value in a probability distribution?**

-> The **expected value (EV)** of a probability distribution is the **average outcome** you would expect over a large number of trials. It represents the **theoretical long-run mean** of a random variable.  

**Formula:**  
For a **discrete variable**:  
$$ E(X) = \sum X_i P(X_i) $$

For a **continuous variable**:  
$$ E(X) = \int x f(x) dx $$

**Example:**  
For a fair 6-sided die:  

$$ E(X) = \sum_{i=1}^{6} i \times \frac{1}{6} = 3.5 $$
This means that over many rolls, the **average outcome will be 3.5**, even though 3.5 is not a possible roll.

**23)  How does a probability distribution relate to the expected outcome of a random variable?**

-> Relationship Between Probability Distribution and Expected Outcome:

The probability distribution defines how likely different outcomes are.
The expected value is the weighted average of all possible outcomes.
Example:
* In gambling, EV helps determine if a game is fair.
* In economics, EV is used for risk analysis.
